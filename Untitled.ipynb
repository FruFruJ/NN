```python
nominalColumns=['Gender','Customer Type','Type of Travel','satisfaction','Class']
```

```python
def showData(data):
    for col in data:
        if(col in nominalColumns):
            sns.countplot(col, data=data)
        else:
            ax=sns.distplot(data[[col]],hist=False,color='pink')
            ax.set(xlabel=col)
        plt.show()
        
```

```python
def nominalToNumeric(data):
    le = LabelEncoder()
    for col in nominalColumns:
        le.fit(data[col])
        data[col]=le.transform(data[col])
    return data
```

```python
dtrain=nominalToNumeric(dtrain)
dtrain.head(5)
```

```python
dtest=nominalToNumeric(dtest)
dtest.head(5)
```

```python
dtest=dtest.dropna(axis=0)
dtest.shape
```

```python
dtrain=dtrain.dropna(axis=0)
dtrain.shape
```

```python
dtest.drop_duplicates().shape
```

```python
dtrain.drop_duplicates().shape
```

```python
def StandardScale(data):
    scale=StandardScaler()
    datapom=pd.DataFrame()
    clas = data['satisfaction']
    datapom = data.loc[:, data.columns != 'satisfaction']
    datapom = pd.DataFrame(scale.fit_transform(datapom.values), columns=datapom.columns, index=data.index)
    datapom['satisfaction'] = clas
    return datapom
```

```python
dtestScaled=StandardScale(dtest)
dtestScaled.head(5)
```

```python
dtrainScaled=StandardScale(dtrain)
dtrainScaled.head(5)
```

```python
X_train=dtrainScaled.loc[:,dtrainScaled.columns!='satisfaction']
Y_train=pd.DataFrame()
Y_train['satisfaction']=dtrainScaled['satisfaction']
X_test=dtestScaled.loc[:,dtestScaled.columns!='satisfaction']
Y_test=pd.DataFrame()
Y_test['satisfaction']=dtestScaled['satisfaction']
```

```python
X_train = torch.from_numpy(X_train.to_numpy()).float()
Y_train = torch.squeeze(torch.from_numpy(Y_train.to_numpy()).float())
X_test = torch.from_numpy(X_test.to_numpy()).float()
Y_test = torch.squeeze(torch.from_numpy(Y_test.to_numpy()).float())
print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
class Net(nn.Module):
  def __init__(self, n_features):
    super(Net, self).__init__()
    self.fc1 = nn.Linear(n_features, 25)
    self.fc2 = nn.Linear(25, 30)
    self.fc3=nn.Linear(30,25)
    self.fc4=nn.Linear(25,10)
    self.fc5 = nn.Linear(10, 1)
  def forward(self, x):
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = F.relu(self.fc3(x))
    x = F.relu(self.fc4(x))
    return torch.sigmoid(self.fc5(x))
```

```python
criterion = nn.BCELoss()
```

```python
net = Net(X_train.shape[1])
optimizer = optim.SGD(net.parameters(), lr=0.001)
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
X_train = X_train.to(device)
Y_train = Y_train.to(device)
X_test = X_test.to(device)
Y_test = Y_test.to(device)
net = net.to(device)
criterion = criterion.to(device)
```

```python
def calculate_accuracy(y_true, y_pred):
  predicted = y_pred.ge(.5).view(-1)
  return (y_true == predicted).sum().float() / len(y_true)
```

```python
def round_tensor(t, decimal_places=3):
  return round(t.item(), decimal_places)
for epoch in range(1000):
    y_pred = net(X_train)
    y_pred = torch.squeeze(y_pred)
    train_loss = criterion(y_pred, Y_train)
    if epoch % 100 == 0:
      train_acc = calculate_accuracy(Y_train, y_pred)
      y_test_pred = net(X_test)
      y_test_pred = torch.squeeze(y_test_pred)
      test_loss = criterion(y_test_pred, Y_test)
      test_acc = calculate_accuracy(Y_test, y_test_pred)
      print(
f'''epoch {epoch}
Train set - loss: {round_tensor(train_loss)}, accuracy: {round_tensor(train_acc)}
Test  set - loss: {round_tensor(test_loss)}, accuracy: {round_tensor(test_acc)}
''')
    optimizer.zero_grad()
    train_loss.backward()
    optimizer.step()
```

```python
classes = ['satisfied', 'neutral']
y_pred = net(X_test)
y_pred = y_pred.ge(.5).view(-1).cpu()
Y_test = Y_test.cpu()
print(classification_report(Y_test, y_pred, target_names=classes))
```

```python
cm = confusion_matrix(Y_test, y_pred)
df_cm = pd.DataFrame(cm, index=classes, columns=classes)
hmap = sns.heatmap(df_cm, annot=True, fmt="d")
hmap.yaxis.set_ticklabels(hmap.yaxis.get_ticklabels(), rotation=0, ha='right')
hmap.xaxis.set_ticklabels(hmap.xaxis.get_ticklabels(), rotation=30, ha='right')
plt.ylabel('True label')
plt.xlabel('Predicted label');
```
